```python
#Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.
```

```python
data = pd.read_sql_query('SELECT title, length, CASE WHEN length <= 60 then "duration low" WHEN length <= 120 then "duration medium" ELSE "duration high" END AS "Ranking duration" FROM sakila.film WHERE length is not null AND length <> 0;', engine)
data.columns = 'Movie', 'Length(mn)', 'Ranking duration'
data
```
